# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" can be treated as different tokens if the text is case-sensitive.  
However, most preprocessing steps convert all text to lowercase, making them the same word.

Example:  
"Love" → "love"

---

## 2. Why are stopwords removed?
Stopwords are common words such as "is", "the", "and" that do not add significant meaning to the sentence.

Removing stopwords helps:
- Reduce noise in the data  
- Improve model performance  
- Focus on meaningful words  

---

## 3. When removing stopwords can be harmful
Removing stopwords can be harmful in tasks like sentiment analysis.

Example:  
"I am not happy"  
If "not" is removed → "I am happy" (wrong meaning)

---

## 4. Difference between stemming and lemmatization

**Stemming:**
- Cuts words to root form  
- Example: "running" → "run"  
- Faster but less accurate  

**Lemmatization:**
- Uses dictionary-based approach  
- Example: "better" → "good"  
- More accurate but slower  

# Task 2:Build Advanced Preprocessing Function

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    important = {"no", "not"}
    tokens = [w for w in tokens if len(w) > 2 or w in important]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

# Task 3: Stress Testing

```python
samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍.",
    "Worst service ever... 0/10.",
    "Call me at 9876543210.",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it.",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in samples:
    tokens, cleaned = preprocess_text(s)
    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*50)
```

# Task 4: Token Analytics

```python
def token_stats(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total

    return total, unique, avg_len


for s in samples:
    tokens, _ = preprocess_text(s)
    total, unique, avg = token_stats(tokens)

    print(s)
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
    print("-"*40)
```

# Task 5: Frequency Analysis

```python
all_tokens = []

for s in samples:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.append(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(samples)
print(result)
```

# Task 7: Error Handling

```python
edge_cases = [
    "",
    "     ",
    "123456",
    "😍😍😍",
    None
]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*40)
```
