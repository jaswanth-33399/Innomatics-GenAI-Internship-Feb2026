# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
  plans = [199, 299, 399, 599]
  if(amount<50):
    return False
  elif(amount in plans):
    return True
  else:
    return False

while True:
  amount = int(input("Enter the recharge amount: "))
  if(validate_recharge(amount)):
    print("Recharge Successful")
    break
  else:
    print("Invalid Recharge Amount please try again\n")
```

# Problem 2: Inventory Reorder Alert System

```python
def inventory_check(stock):
  for i,j in stock.items():
    if j<15:
      print(i,"→ Reorder Alert")
    else:
      print(i,"→ Stock OK")

stock = {
    "rice":10,
    "salt":20,
    "sugar":15,
    "oil":5
}
inventory_check(stock)
```

# Problem 3: Student Result Processing System

```python
def marks_students(marks):
  total = 0
  for i in marks:
    total += i
  avg = total/len(marks)
  print("Average Marks: ",avg)
  if(avg<50):
    return "Fail"
  else:
    return "Pass"

marks = [50,60,70,80,90]
result = marks_students(marks)
print("result: ",result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def fare(dist,peak):
  base = 50
  total = base+(dist*12)
  if peak=="Yes":
    total += total*0.25
  return total

while(True):
  dist = int(input("Enter the distance: "))
  peak = input("Is the journey peak time? (Yes/No): ")
  amount = fare(dist,peak)
  print("The fare is: ",amount)
  retry = input("Do you want to retry? (Yes/No): ")
  if retry=="No":
    break
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def attendence(atten):
  count = 0
  for i in atten:
    if i=="P":
      count += 1
  Percentage = (count/len(atten))*100
  if(Percentage>=75):
    return "Eligible"
  else:
    return "Not Eligible"

atten = ["P", "P", "A", "P", "P"]
result = attendence(atten)
print("Status: ",result)
```

# Problem 6: Password Strength Checker

```python
def password(Pass):
  has_digit = False
  has_special = False
  special = "@#$"
  if len(Pass)<8:
    return "Weak Password (Minimum length should be 8)"
  for i in Pass:
    if i.isdigit():
      has_digit = True
    if i in special:
      has_special = True
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password (Must contain digit and special char)"

Pass = input("Enter the password: ")
result = password(Pass)
print(result)
```
