# Problem 1: Unique Words in a Sentence

```python
sentence = "python is easy and python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words count: ",len(unique_words))
print("Unique words: ",unique_words)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
high_emp = max(employees, key=employees.get)
print(f"Highest Salary: {high_emp} - {employees[high_emp]}")
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
minimum = numbers[0]
maximum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Minimum:", minimum)
print("Maximum:", maximum)
```

# Problem 4: Count Products Above a Price Threshold



```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
  if i>1000:
    count += 1
print("Products above 1000: ",count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count = 0
for i in attendance:
  if i=="P":
    count += 1
print("Attendance Percentage: ",count/len(attendance)*100)
```

# Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
phone_numbers = set(phone_numbers)
print("Unique Phone Numbers: ",phone_numbers)
```

# Problem 7: Count Character Frequency

```python
string = input()
freq = {}
for i in string:
  freq[i] = freq.get(i,0)+1
print(freq)
```
