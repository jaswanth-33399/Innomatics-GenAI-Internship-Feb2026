# Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
  "Ravi": 92,
  "Anita": 88,
  "Kiran": 92,
  "Suresh": 85
}
high = max(employees.values())
top = []
for i,j in employees.items():
  if(high == j):
    top.append(i)
print(f"Top Performers Eligible for Bonus: {', '.join(top)} (Score: {high})")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
from typing import Counter
import string
query = "Buy mobile phone buy phone online"
query = query.lower()
for i in query:
  if i==string.punctuation:
    query.remove(i)
words = query.split()
freq = {}
for i in words:
  freq[i] = freq.get(i,0)+1
res = {}
for i,j in freq.items():
  if j>1:
    res[i] = j
print(res)
```

# Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
mail = {}
for i in emails:
  j = i.split("@")[1]
  mail[j] = mail.get(j,0)+1
tot = len(emails)
for i, j in mail.items():
    percentage = (j / tot) * 100
    print(f"{i}: {int(percentage)}%")
```

# Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
freq = {}
for i in user_ids:
  freq[i] = freq.get(i,0)+1
for i,j in freq.items():
  if j>1:
    print(f"{i} → {j} times")
```
