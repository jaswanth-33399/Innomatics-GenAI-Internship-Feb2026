# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
n = int(input("Enter number of entries: "))
likes = []
print("Enter likes:")
for _ in range(n):
    likes.append(int(input()))
analyze_engagement(likes)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_adequacy_checker(rainfall_data, required_level):
    total_rainfall = 0
    for rain in rainfall_data:
        total_rainfall += rain
    average_rainfall = total_rainfall // len(rainfall_data)
    print("Average Rainfall:", average_rainfall)
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
seasonal_rainfall = [70, 75, 68, 74, 73]  # average = 72
required_rainfall = 70
rainfall_adequacy_checker(seasonal_rainfall, required_rainfall)
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_users = set()
    duplicates_found = False
    for user in usernames:
        if user in unique_users:
            duplicates_found = True
            break
        unique_users.add(user)
    if duplicates_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
user_list = ["alex", "john", "maria", "alex", "steve"]
detect_duplicate_accounts(user_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_price_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crop_price_filter(crop_prices)
```

# Problem 9: E-Learning – Student Attendance Eligibility

```python
def attendance_eligibility_checker(attendance_records):
    total_classes = len(attendance_records)
    attended_classes = 0
    for day in attendance_records:
        if day == 1:
            attended_classes += 1

    attendance_percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", attendance_percentage)

    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance = [1, 1, 0, 1, 1]
attendance_eligibility_checker(attendance)
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_analyzer(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage = "Low Usage"
    elif bill < 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage
units_consumed = 300
total_bill, usage_status = electricity_bill_analyzer(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
